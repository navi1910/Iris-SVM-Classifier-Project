# SVM - Iris - Classifier Project
## To classify the oberservations into the right Target Labels.

#### The iris dataset contains measurements for 150 iris flowers from three different species.

#### The three classes in the Iris dataset:

    Iris-setosa (n=50)
    Iris-versicolor (n=50)
    Iris-virginica (n=50)

##### The four features of the Iris dataset:

    sepal length in cm
    sepal width in cm
    petal length in cm
    petal width in cm

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# Loading the data as a data-frame
df = sns.load_dataset('iris')
```

```python
# Checking basic information about the data.
df.head()
```

## Exploratory Data Analysis

```python
sns.countplot(df['species'])
```

```python
sns.set_style('whitegrid')
sns.pairplot(df, hue='species', palette='Dark2')
```

```python
sns.set_style('white')
plt.figure(figsize=(6,5))
sns.kdeplot(x = 'sepal_length', y = 'sepal_width', data = df[df['species'] == 'setosa'], kind ='kde',
             cmap="plasma", shade=True, )
```

## Model Building

```python
X = df.drop('species', axis = 1)
y = df['species']
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 1)
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
confusion_matrix(y_test, pred)
```

```python
params = {'C':[0.1, 1, 10, 100, 1000], 'gamma':[10, 1, 0.1, 0.01, 0.001]}
```

```python
gs = GridSearchCV(model, params)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
gs.best_estimator_
```

```python
gs.best_score_
```

```python
predictions = gs.predict(X_test)
```

## Results

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
print('Accuracy: ', accuracy_score(y_test, predictions))
```
